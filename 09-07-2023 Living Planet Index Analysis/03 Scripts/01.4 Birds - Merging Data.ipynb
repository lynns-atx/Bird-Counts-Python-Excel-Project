## 2. Adding back in columns on taxonomy (order, family, genus, species) by merging two tables

```python
#Joining order, family, genus and species onto birds dataframe
birds_merge = birds_clean.merge(birds_tax, on = ['id', 'scientific_name', 'class', 'common_name', 'country', 'region', 'system'], indicator = True)
```
